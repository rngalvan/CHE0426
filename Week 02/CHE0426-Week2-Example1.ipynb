```python
x0 = 10
fx = 100
iteration = 0
ff = []
xx = []

while abs(fx) > 1.0e-5:
    # Calculate fx
    fx = 2 * (10*x0**3 + 3*x0**2 + x0 + 5) * (30*x0**2 + 6*x0 + 1)

    # Append to list
    ff.append(fx)

    # Calculate fxp
    fxp = 2 * ((30*x0**2 + 6*x0 + 1) * (30*x0**2 + 6*x0 + 1) +
               (10*x0**3 + 3*x0**2 + 5) * (60*x0 + 6))

    # Newton-Raphson update
    x0 = x0 - fx / fxp

    # Append to list
    xx.append(x0)

    iteration += 1

# Print results
print(f"fx   = {fx:.4e}")
print(f"iter = {iteration}")
print(f"x0   = {x0:.4f}")
```
