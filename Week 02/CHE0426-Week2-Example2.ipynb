```python
x0 = 10
fx = 100
iteration = 0
ff = []
xx = []
dx = 0.001

while abs(fx) > 1.0e-5:
    # Calculate fx
    fx = 2 * (10*x0**3 + 3*x0**2 + x0 + 5) * (30*x0**2 + 6*x0 + 1)

    # Append fx to list
    ff.append(fx)

    # Calculate perturbed x
    xp = x0 + dx

    # Calculate f(x + dx)
    ffp = 2 * (10*xp**3 + 3*xp**2 + x0 + 5) * (30*xp**2 + 6*xp + 1)

    # Numerical derivative (Finite Difference)
    fxp = (ffp - fx) / dx

    # Newton-Raphson update
    x0 = x0 - fx / fxp

    # Append x0 to list
    xx.append(x0)

    iteration += 1

# Print results
print(f"fx   = {fx:.4e}")
print(f"iter = {iteration}")
print(f"x0   = {x0:.4f}")
```
