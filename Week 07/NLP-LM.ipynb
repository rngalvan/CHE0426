```
import numpy as np
from scipy.optimize import minimize

def optimize_pcm_cylinder(target_surface_area):
    """
    Finds the optimal radius and height to maximize volume
    for a given fixed surface area.
    """

    # 1. Objective Function: MINIMIZE negative volume
    # x[0] is radius (r), x[1] is height (h)
    def objective(x):
        r, h = x
        volume = np.pi * (r**2) * h
        return -volume  # Negative sign to maximize

    # 2. Constraint Function: Surface area must equal target_surface_area
    # scipy expects constraints in the format: function(x) == 0
    def constraint(x):
        r, h = x
        surface_area = 2 * np.pi * (r**2) + 2 * np.pi * r * h
        return surface_area - target_surface_area

    # Define the constraint dictionary for scipy
    cons = {'type': 'eq', 'fun': constraint}

    # 3. Bounds: Radius and height must be strictly positive
    # (r_min, r_max), (h_min, h_max)
    bounds = ((0.01, None), (0.01, None))

    # 4. Initial Guess: Starting point for the algorithm (r=1, h=1)
    initial_guess = [1.0, 1.0]

    # 5. Run the Optimization
    result = minimize(
        objective,
        initial_guess,
        method='SLSQP',
        bounds=bounds,
        constraints=cons
    )

    # Extract results
    optimal_r = result.x[0]
    optimal_h = result.x[1]
    max_volume = -result.fun # Reverse the negative sign to get actual volume

    return optimal_r, optimal_h, max_volume

# --- Run the Simulation ---
A_0 = 100.0  # Let's set a strictly fixed surface area of 100 square units
r_opt, h_opt, v_max = optimize_pcm_cylinder(A_0)

print(f"--- Optimization Results for Surface Area = {A_0} ---")
print(f"Optimal Radius (r): {r_opt:.4f}")
print(f"Optimal Height (h): {h_opt:.4f}")
print(f"Maximum Volume (V): {v_max:.4f}")
print("-" * 45)

# The Computational Proof of our Math
ratio = h_opt / r_opt
print(f"Ratio of Height to Radius (h/r): {ratio:.4f}")
print(f"Does h = 2r? {'Yes' if np.isclose(ratio, 2.0) else 'No'}")
```
