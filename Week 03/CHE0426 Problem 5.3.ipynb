```python
import pandas as pd
import numpy as np

# 1. Define the function and its derivatives
def f(x):
    return (x - 1)**4

def df(x):
    return 4 * (x - 1)**3

def d2f(x):
    return 12 * (x - 1)**2

# 2. Implement Newton's Method
def newtons_method(x0, tol=1e-6, max_iter=50):
    x = x0
    history = [x]

    print(f"\n--- Newton's Method (Start: {x0}) ---")
    print(f"{'Iter':<5} {'x_value':<12} {'f(x)':<12} {'f\'(x)':<12}")

    for i in range(max_iter):
        first_deriv = df(x)
        second_deriv = d2f(x)

        print(f"{i:<5} {x:<12.6f} {f(x):<12.6f} {first_deriv:<12.6f}")

        if abs(first_deriv) < tol:
            break

        if second_deriv == 0:
            print("Zero division error in second derivative.")
            break

        # Update rule
        x = x - first_deriv / second_deriv
        history.append(x)

    return x

# 3. Implement Quasi-Newton (Secant) Method
def secant_method(x_current, x_prev, tol=1e-6, max_iter=50):
    history = [x_current]

    print(f"\n--- Secant Method (Start: {x_current}) ---")
    print(f"{'Iter':<5} {'x_value':<12} {'f(x)':<12} {'f\'(x)':<12}")

    for i in range(max_iter):
        df_curr = df(x_current)
        df_prev = df(x_prev)

        print(f"{i:<5} {x_current:<12.6f} {f(x_current):<12.6f} {df_curr:<12.6f}")

        if abs(df_curr) < tol:
            break

        if (df_curr - df_prev) == 0:
            print("Division by zero in secant slope.")
            break

        # Update rule
        # New x = current x - f'(x) * (change in x) / (change in f')
        x_new = x_current - df_curr * (x_current - x_prev) / (df_curr - df_prev)

        x_prev = x_current
        x_current = x_new
        history.append(x_current)

    return x_current

# --- Execution ---
starting_points = [-1, -0.5, 0.0]

for start_x in starting_points:
    print(f"\n" + "="*40)
    print(f"Solving for Starting Point x = {start_x}")
    print("="*40)

    # Run Newton's Method
    newtons_method(start_x)

    # Run Secant Method
    # Assumption: We generate a previous point by subtracting a small delta (0.1)
    # to kickstart the secant process.
    secant_method(start_x, start_x - 0.1)
```
