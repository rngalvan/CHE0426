```python
import numpy as np
import matplotlib.pyplot as plt

def bracket_minimum(func, start_x=0.0, initial_step=0.1, max_iter=100):
    """
    Bracketing Method (Bounding Phase).
    Searches for a range [a, c] that contains a local minimum.
    """
    k = 0
    x1 = start_x
    x2 = x1 + initial_step

    f1 = func(x1)
    f2 = func(x2)

    # 1. Determine initial direction.
    # If f2 > f1, we are going uphill. Reverse direction.
    if f2 > f1:
        initial_step = -initial_step
        x2 = x1 + initial_step
        f2 = func(x2)
        # If it's still higher, the minimum is bracketed between x2 and (x1 + |old_step|)
        if f2 > f1:
             return (x2, x1, x1 - initial_step), 1

    # 2. Iterate with step expansion (acceleration)
    step = initial_step

    while k < max_iter:
        k += 1
        step = step * 2.0  # Accelerate step size
        x3 = x2 + step
        f3 = func(x3)

        # Check if function started rising
        if f3 > f2:
            # Minimum is bracketed!
            # Sort the points so a < b < c
            points = sorted([x1, x2, x3])
            return tuple(points), k

        # Move forward
        x1 = x2
        f1 = f2
        x2 = x3
        f2 = f3

    return None, k

# --- Define the Functions from the Image ---
def func_a(x): return np.exp(x) + 1.5 * x**2
def func_b(x): return 0.5 * (x**2 + 1) * (x + 1)
def func_c(x): return x**3 - 3*x
def func_d(x): return 2 * x**2 * (x - 2) * (x + 2)
def func_e(x): return (0.1*x**6 - 0.29*x**5 + 2.31*x**4 -
                       8.33*x**3 + 12.89*x**2 - 6.8*x + 1)

# List of functions for the loop
problems = [
    ("a", func_a),
    ("b", func_b),
    ("c", func_c),
    ("d", func_d),
    ("e", func_e)
]

print(f"{'Func':<5} | {'Bracket Found [a, c]':<25} | {'Iterations':<5} | {'Note'}")
print("-" * 65)

for label, f in problems:
    bracket, iters = bracket_minimum(f, start_x=0, initial_step=0.1)

    if bracket:
        a, b, c = bracket
        note = "Success"
        print(f"({label})   | [{a:.4f}, {c:.4f}]             | {iters:<5}      | {note}")
    else:
        print(f"({label})   | {'FAILED to bracket':<25} | {iters:<5}      | Unbounded?")
```
