```python
def f(x):
    """The objective function to minimize."""
    return (x - 1)**4

def quadratic_interpolation_search():
    # 1. Initialization
    # Initial points given in the problem
    points = [0.0, 0.5, 2.0]

    # Track function evaluations
    # We evaluate the initial 3 points first
    eval_count = 0

    # Store points as dictionaries for easy sorting: {'x': value, 'y': f(value)}
    current_points = []
    for x_val in points:
        y_val = f(x_val)
        current_points.append({'x': x_val, 'y': y_val})
        eval_count += 1

    print(f"{'Iter':<5} {'x1':<10} {'x2':<10} {'x3':<10} {'f(x1)':<10} {'f(x2)':<10} {'f(x3)':<10} {'x_new':<10} {'f(x_new)':<10}")
    print("-" * 100)

    iteration = 0
    max_evals = 10

    # Loop until we hit the evaluation limit
    # We need at least 1 remaining evaluation to calculate f(x_new)
    while eval_count < max_evals:
        iteration += 1

        # Sort points by x-coordinate for the formula logic (x1 < x2 < x3)
        # (Though the formula works regardless of order, it's good for display)
        current_points.sort(key=lambda p: p['x'])

        x1, x2, x3 = current_points[0]['x'], current_points[1]['x'], current_points[2]['x']
        f1, f2, f3 = current_points[0]['y'], current_points[1]['y'], current_points[2]['y']

        # 2. Calculate the minimum of the parabola passing through these 3 points
        # Formula: x* = 0.5 * [ (x2^2 - x3^2)f1 + (x3^2 - x1^2)f2 + (x1^2 - x2^2)f3 ] / [ (x2 - x3)f1 + (x3 - x1)f2 + (x1 - x2)f3 ]

        numerator = (x2**2 - x3**2) * f1 + (x3**2 - x1**2) * f2 + (x1**2 - x2**2) * f3
        denominator = 2 * ((x2 - x3) * f1 + (x3 - x1) * f2 + (x1 - x2) * f3)

        if denominator == 0:
            print("Denominator is zero (collinear points). Cannot proceed.")
            break

        x_new = numerator / denominator

        # 3. Evaluate the new point
        f_new = f(x_new)
        eval_count += 1

        # Print current state
        print(f"{iteration:<5} {x1:<10.4f} {x2:<10.4f} {x3:<10.4f} {f1:<10.4f} {f2:<10.4f} {f3:<10.4f} {x_new:<10.4f} {f_new:<10.4f}")

        # 4. Update the set of points
        # We now have 4 points. We keep the 3 best (lowest f(x)) to zoom in on the minimum.
        # Create a new candidate list
        candidates = current_points + [{'x': x_new, 'y': f_new}]

        # Sort by function value (y) ascending
        candidates.sort(key=lambda p: p['y'])

        # Keep the top 3
        current_points = candidates[:3]

    # Final Result
    # The best estimate is the point with the lowest function value in our final set
    best_point = current_points[0] # Since we just sorted them

    print("-" * 100)
    print(f"Total Function Evaluations: {eval_count}")
    print(f"Best estimate for minimum x: {best_point['x']:.6f}")
    print(f"Minimum f(x): {best_point['y']:.6f}")
    print(f"True minimum is at x = 1.0")

# Run the solution
if __name__ == "__main__":
    quadratic_interpolation_search()
```
