```python
import math

def f(x):
    """The function to maximize."""
    return 6.64 + 1.2 * x - x**2

def golden_section_search_max(a, b, tolerance_percent):
    """
    Performs Golden Section Search to find the maximum of f(x).

    Args:
        a (float): Lower bound of interval.
        b (float): Upper bound of interval.
        tolerance_percent (float): Desired final interval size as a fraction (e.g., 0.02).
    """

    # 1. Setup Constants
    # Golden Ratio conjugate (rho = 0.61803...)
    rho = (math.sqrt(5) - 1) / 2

    original_uncertainty = b - a
    target_uncertainty = tolerance_percent * original_uncertainty

    # 2. Initialize points x1 (left) and x2 (right)
    # x1 corresponds to distance rho from the right (or 1-rho from left)
    # x2 corresponds to distance rho from the left

    x1 = b - rho * (b - a)
    x2 = a + rho * (b - a)

    f1 = f(x1)
    f2 = f(x2)

    iteration = 0

    # Header for the iteration table
    print(f"{'Iter':<5} {'a':<10} {'b':<10} {'Uncertainty':<12} {'x1':<10} {'x2':<10} {'f(x1)':<10} {'f(x2)':<10}")
    print("-" * 85)

    # 3. Iteration Loop
    current_uncertainty = b - a
    while current_uncertainty > target_uncertainty:
        iteration += 1

        # Print current state
        print(f"{iteration:<5} {a:<10.5f} {b:<10.5f} {current_uncertainty:<12.5f} {x1:<10.5f} {x2:<10.5f} {f1:<10.5f} {f2:<10.5f}")

        # Update Logic for MAXIMIZATION
        if f1 < f2:
            # The maximum is in the right sub-interval [x1, b].
            # We discard the left part [a, x1].
            a = x1

            # The old x2 becomes the new x1
            x1 = x2
            f1 = f2

            # Calculate new x2
            x2 = a + rho * (b - a)
            f2 = f(x2)
        else:
            # The maximum is in the left sub-interval [a, x2].
            # We discard the right part [x2, b].
            b = x2

            # The old x1 becomes the new x2
            x2 = x1
            f2 = f1

            # Calculate new x1
            x1 = b - rho * (b - a)
            f1 = f(x1)

        current_uncertainty = b - a

    # 4. Final Result
    print("-" * 85)
    print(f"Stop condition met: Uncertainty {current_uncertainty:.5f} < {target_uncertainty}")
    print(f"Final Interval: [{a:.5f}, {b:.5f}]")
    best_estimate = (a + b) / 2
    print(f"Estimated Max x: {best_estimate:.5f}")
    print(f"Max f(x): {f(best_estimate):.5f}")

# --- Execute the Solver ---
# Interval [0, 1], reduce to < 2% (0.02)
golden_section_search_max(0.0, 1.0, 0.02)
```
