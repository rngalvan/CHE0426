```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.exp(x) - 1.5 * x**2

def attempt_bracket(start_x, step_size):
    x0 = start_x
    x1 = x0 + step_size
    f0 = f(x0)
    f1 = f(x1)

    print(f"\n--- Attempting bracket with step size: {step_size} ---")
    print(f"Start: x={x0}, f(x)={f0:.4f}")
    print(f"Step : x={x1}, f(x)={f1:.4f}")

    if f1 < f0:
        print("Function is decreasing. Continuing search...")

        if step_size < 0:
            print("RESULT: Function decreases towards negative infinity. No minimum bracketed.")
        else:
            print("RESULT: Found a descent direction.")
    else:
        print("Function is increasing immediately.")
        print("Standard algorithms assume the minimum is in the OTHER direction.")
        print("RESULT: Failed to bracket minimum in this direction.")

# --- 1. Analyze with different Step Sizes ---
attempt_bracket(0, 0.1)  # Small Step (Right)
attempt_bracket(0, 2.0)  # Large Step (Right)
attempt_bracket(0, -1.0) # Step Left

# --- 2. Plot the function to explain the results ---
x_vals = np.linspace(-2, 2.5, 100)
y_vals = f(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='$f(x) = e^x - 1.5x^2$', color='blue')
plt.plot(0, f(0), 'ro', label='Start (x=0)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Plot of $f(x) = e^x - 1.5x^2$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.ylim(-5, 5)
plt.show()
```
