```python
import numpy as np

def f(x):
    return 1 - 8*x + 2*x**2 - (10/3)*x**3 + 0.25*x**4 + (4/5)*x**5 - (1/6)*x**6

def df(x):
    # Provided hint: f'(x) = (1+x)^2 * (2-x)^3
    return ((1 + x)**2) * ((2 - x)**3)

def d2f(x):
    # Derived above: f''(x) = (1+x)(2-x)^2 * (1-5x)
    return (1 + x) * ((2 - x)**2) * (1 - 5 * x)

# --- Part (a): Analytical Check ---
def part_a_analytical():
    print("\n--- (a) Analytical Solution ---")
    print("Critical points are at x = -1 and x = 2.")
    print(f"f(-1) = {f(-1):.4f}")
    print(f"f(2)  = {f(2):.4f}")

    if f(2) > f(-1):
        print("Global Maximum is at x = 2")
    else:
        print("Global Maximum is at x = -1")

# --- Part (b): Newton's Method ---
def part_b_newton():
    print("\n--- (b) Newton's Method ---")
    print(f"{'Iter':<5} {'x_curr':<10} {'f\'(x)':<12} {'f\'\'(x)':<12} {'x_next':<10}")
    print("-" * 60)

    x = -2.0 # Starting point
    for i in range(1, 3): # 2 iterations requested
        first = df(x)
        second = d2f(x)

        if second == 0:
            print("Error: Second derivative is zero.")
            break

        x_next = x - (first / second)

        print(f"{i:<5} {x:<10.4f} {first:<12.4f} {second:<12.4f} {x_next:<10.4f}")
        x = x_next

    print(f"Result after 2 iterations: x = {x:.4f}")

# --- Part (c): Quadratic Interpolation ---
def part_c_quadratic():
    print("\n--- (c) Quadratic Interpolation ---")
    # Since only one start point is given, we assume an initial step size of 1.0
    # to create the necessary 3 points.
    x0, x1, x2 = -2.0, -1.0, 0.0

    print(f"Initial points chosen: {x0}, {x1}, {x2}")
    print(f"{'Iter':<5} {'x0':<8} {'x1':<8} {'x2':<8} {'f(x0)':<8} {'f(x1)':<8} {'f(x2)':<8} {'x_new':<8}")
    print("-" * 80)

    for i in range(1, 3): # 2 iterations requested
        f0, f1, f2 = f(x0), f(x1), f(x2)

        # Quadratic peak formula
        numerator = (x1**2 - x2**2)*f0 + (x2**2 - x0**2)*f1 + (x0**2 - x1**2)*f2
        denominator = 2 * ((x1 - x2)*f0 + (x2 - x0)*f1 + (x0 - x1)*f2)

        if denominator == 0:
            print("Denominator zero in quadratic interpolation.")
            break

        x_new = numerator / denominator
        f_new = f(x_new)

        print(f"{i:<5} {x0:<8.4f} {x1:<8.4f} {x2:<8.4f} {f0:<8.4f} {f1:<8.4f} {f2:<8.4f} {x_new:<8.4f}")

        # Logic to choose the best 3 points for the next iteration (Maximization)
        # We sort all 4 points by function value and keep the top 3
        points = [
            {'x': x0, 'v': f0},
            {'x': x1, 'v': f1},
            {'x': x2, 'v': f2},
            {'x': x_new, 'v': f_new}
        ]
        # Sort descending because we want the MAXIMUM
        points.sort(key=lambda p: p['v'], reverse=True)

        # Reassign x0, x1, x2 for next loop based on best values
        # Note: Standard algo often maintains order x0<x1<x2, but for pure maximization
        # finding the peak, keeping the highest values is the robust strategy.
        x0 = points[0]['x']
        x1 = points[1]['x']
        x2 = points[2]['x']

    print(f"Best estimate after 2 iterations: x = {points[0]['x']:.4f}")

# --- Run All ---
part_a_analytical()
part_b_newton()
part_c_quadratic()
```
